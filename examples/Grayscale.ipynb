```python
import cv2
import matplotlib.pylab as plt
from plakakia.utils_tiling import tile_image
```

```python
im_gray = cv2.imread('images/grayscale_dog.jpeg', cv2.IMREAD_GRAYSCALE)
print(f"The shape of the image is {im_gray.shape}")
```

```python
plt.imshow(im_gray,cmap="gray")
plt.colorbar();
```

```python
tiles, coordinates = tile_image(im_gray, tile_size=100, step_size=100)
```

```python
var = int(len(tiles)**(1/2))

import matplotlib.pyplot as plt
fig, ax = plt.subplots(var, var, figsize=(5, 5))
# Add a super title
fig.suptitle('Tiles of the original image', fontsize=20)
for i in range(var):
    for j in range(var):
        ax[i, j].imshow(tiles[i*var+j],cmap="gray")
        ax[i, j].axis('off')
plt.show()
```
